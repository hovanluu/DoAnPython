```python
# lấy phần tháng của dataframe ra ngoài bằng phương thức: Series.str.slice(start=None, stop=None, step=None)
df1 = df['Order Date'].str.slice(start=0, stop=2, step=1)
df1
```

```python
#thêm cột tháng vào dataframe 
df['Month'] = df1
# hiện 5 dòng đầu 
df.head()
```

```python
# Kiểm tra xem dataframe có giá trị nan và Or không ? 
print(set(df['Month']))
```

```python
#Loại bỏ giá trị nan
df = df.dropna(how='all')
df.head()
```

```python
#Loại bỏ giá trị Or
df = df[df['Month'] != 'Or']
df.head()
```

```python
#Tháng bán hàng tốt nhất ? Tháng đó kiếm được bao nhiêu ? 
#Tạo thêm cột sales để tính giá trị từng đơn hàng bằng cách sales  = ['Quantity Ordered'] * ['Price Each']
#Kiểm tra xem 2 cột ['Quantity Ordered'] , ['Price Each'] có phải là kiểu dữ liệu số hay không ?
print(df['Quantity Ordered'].dtypes)
print (df['Price Each'].dtypes)
```

```python
# vì ['Quantity Ordered'] là kiểu dữ liệu int nên ta phải chuyển thành kiểu int  bằng phương thức : pd.to_numeric(arr,downcast)
# vì ['Price Each'] là kiểu dữ liệu int nên ta phải chuyển thành kiểu float  bằng phương thức : pd.to_numeric(arr,downcast)
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast='integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast='float')
#sau khi chuyển đổi kiểm tra lại kiểu dữ liệu của 2 cột
print(df['Quantity Ordered'].dtypes)
print (df['Price Each'].dtypes)
#tiến hành thêm cột sales để tính giá trị từng đơn hàng bằng cách sales  = ['Quantity Ordered'] * ['Price Each']
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

```python
#lọc dữ liệu theo từng tháng 
df.groupby('Month').sum()
```

```python
#Vì mình chỉ cần giá trị của cột sale thôi nên ta sẽ chỉ lấy cột sale 
df.groupby('Month').sum()['Sales']
```

```python
#Tìm tháng có doanh số lớn nhất
sales_value = df.groupby('Month').sum()['Sales']
sales_value.max()
```

```python
#Tiếp theo mình sẽ tải dữ liệu vào biểu đồ 
#Cài đặt và Import thư viện matplotlib
months = range(1,13)
plt.bar(x= months, height = sales_value)
plt.show()
```

```python
#do trục x bị mất những tháng lẻ nên ta sẽ dùng phương thức plt.xticks(truyền vào tham số mình muốn thể hiện)
#cho trục x và trục y 1 cái tên để người đọc hiểu được ý nghĩa của nó bằng phương thức : plt.label('')
months = range(1,13)
plt.bar(x= months, height = sales_value)
plt.xticks(months)
plt.xlabel('Tháng')
plt.ylabel('Bán Hàng')
plt.show()
```

```python
#Thành phố có doanh thu bán hàng tốt nhất.
#Tách thành phố từ Purchase Address
#Dùng phương split để tách chuỗi kí tự bằng dấu phẩy 
address_to_city = lambda address:address.split(',')[1]
df['City'] = df['Purchase Address'].apply(address_to_city)
df.head()
```

```python
df.groupby('City')['Sales'].sum()
```

```python
Sales_city = df.groupby('City')['Sales'].sum()
Sales_city.max()
```

```python
# Thể hiện dưới dạng biểu đồ 
City = range(1,10)
plt.bar(x= City,height =  Sales_city)
plt.show()
```

```python
# Thể hiện dưới dạng biểu đồ 
plt.bar(x= City,height =  Sales_city)
plt.xticks(City)
plt.xlabel('Thành Phố')
plt.ylabel('Doanh số của thành Phố')
plt.show()
```
